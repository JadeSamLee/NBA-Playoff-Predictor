# Exploratory Data Analysis (EDA)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
nba_data = pd.read_csv('/content/nba_team_data.csv')
```

```python
# Display basic information and summary statistics
print(nba_data.info())
print(nba_data.describe())
```

```python
# Check for missing values
print(nba_data.isnull().sum())
```

```python
# Visualize the distribution of wins and losses
plt.figure(figsize=(12, 6))
sns.histplot(nba_data['W'], bins=30, kde=True)
plt.title('Distribution of Wins')
plt.xlabel('Wins')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(nba_data['L'], bins=30, kde=True)
plt.title('Distribution of Losses')
plt.xlabel('Losses')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualize win/loss percentage
plt.figure(figsize=(12, 6))
sns.histplot(nba_data['W/L%'].astype(float), bins=30, kde=True)
plt.title('Distribution of Win/Loss Percentage')
plt.xlabel('Win/Loss Percentage')
plt.ylabel('Frequency')
plt.show()
```

```python
# Check playoff appearance by division
plt.figure(figsize=(14, 7))
sns.countplot(data=nba_data, x='Playoffs', hue='W/L%')
plt.title('Playoff Appearance by Win/Loss Percentage')
plt.xlabel('Playoffs')
plt.ylabel('Count')
plt.show()
```
